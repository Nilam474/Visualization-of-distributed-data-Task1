```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# To display plots inlin
%matplotlib inline

#load dataset
file_path = r"C:\Users\Nilam\Downloads\API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv"
df = pd.read_csv(file_path, skiprows=4)

print("Dataset loaded successfully!\n")
print(df.head())

# Basic info
print("\nDataset Info:")
print(df.info())

df = df.dropna(axis=1, how='all')

# Visualize population of recent year (say 2020)
if '2020' in df.columns:
    plt.figure(figsize=(10,6))
    top_countries = df[['Country Name', '2020']].sort_values(by='2020', ascending=False).head(10)
    sns.barplot(x='2020', y='Country Name', data=top_countries, palette='crest')
    plt.title('Top 10 Most Populated Countries (2020)')
    plt.xlabel('Population')
    plt.ylabel('Country')
    plt.show()
else:
    print("Year column 2020 not found!")

# Histogram â€” distribution of 2020 population
if '2020' in df.columns:
    plt.figure(figsize=(8,5))
    sns.histplot(df['2020'].dropna(), bins=30, kde=True, color='orange')
    plt.title('Distribution of Country Populations (2020)')
    plt.xlabel('Population')
    plt.ylabel('Number of Countries')
    plt.show()
else:
    print("Column 2020 not found for histogram.")
```
